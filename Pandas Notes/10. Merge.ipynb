```python
import pandas as pd
import numpy as np
```

```python
d = {
    'NO':[11,22,33,88,44,55,88,99],
    'ID': [1,2,3,4,8,5,8,9],
    'NAME': [111,888,222,333,444,555,888,999],
    'SCORE': [1111,8888,2222,3333,4444,5555,8888,9999],
}

dd = {
    'ID':[1,2,6,3,4,5,6,7],
    'NO1': [10,20,50,30,40,50,50,60],
    'NAME1': [100,200,300,500,400,500,500,600],
    'SCORE1': [1000,2000,3000,5000,4000,5000,5000,6000],
}
```

```python
df1 = pd.DataFrame(data=d)
df2 = pd.DataFrame(data=dd)
```

```python
print(df1,df2, end=" ")
```

```python
pd.merge(df1,df2, on='ID')
```

```python
pd.merge(df1,df2, how='inner')
```

```python
pd.merge(df1,df2, how='outer')
```

```python
pd.merge(df1,df2, how='outer', indicator=True)
```

```python
pd.merge(df1,df2,how='outer', left_index=True, right_index=True)
```

```python
pd.merge(df1,df2,how='outer', left_index=True, right_index=True, suffixes=('_R', '_L'))
```
