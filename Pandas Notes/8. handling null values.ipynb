```python
import pandas as pd
import numpy as np
```

```python
d = {
    'ID': [np.nan,200,300,400,500,600,700,800,900,np.nan,np.nan],
    'NAME': ['ravi','rajesh', 'saru', 'nilu', np.nan, 'sihvani', 'radha', 'amar', 'saku', np.nan,np.nan],
    'SCORE': [10,20,30,np.nan,50,60,70,80,90,np.nan,np.nan]
}
```

```python
df = pd.DataFrame(data=d)
```

```python
# using  perticular subset we delete the null and rows according to that subset
df.dropna(subset=["SCORE"])
```

```python
# fill null alues with respect to the attributes
df.fillna({
    'ID': "RollNumber",
    'NAME': "Name of the person",
    "SCORE": "Marks",
})
```
